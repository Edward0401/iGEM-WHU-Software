```python
import csv
import os
import pandas as pd
```

```python
def getScoreFromBlast():
    BlastResult5=pd.read_csv('out.tsv', sep='\t',header=None)
    score=0
    for i in range(0,len(BlastResult5)):
        score=score+BlastResult5.iloc[i,11]/10
    return score
```

```python
Seq_csv='combinedsequence10bp.csv'
```

```python
Sequences=pd.read_csv(Seq_csv, sep=',',header=None)
```

```python
Sequences
```

```python
Sequences.insert(Sequences.shape[1], 'd', 0)
```

```python
for i in range(0,len(Sequences)):
        seq=Sequences.iloc[i,0]
        print(seq,str(i))
        with open('in.blasta', 'w') as f:
                f.write('>Seq'+str(i))
                f.write('\n')
                f.write(seq)
                f.close()
        os.system(" blastn -query .\in.blasta -db ecoli -task blastn-short -word_size 4 -evalue 1000 -outfmt 6 -out out.tsv")
        bitscore=getScoreFromBlast()
        print(bitscore)
        os.remove('out.tsv')
        os.remove('in.blasta')
        Sequences.iloc[i,1]=bitscore
        
        
                
```
